```python
# Load tokenizer for SpeechT5
from transformers import SpeechT5Tokenizer

tokenizer = SpeechT5Tokenizer.from_pretrained("microsoft/speecht5_tts")

# function to extract all the characters from the dataset 
def extract_all_chars(batch):
    all_text = " ".join(batch["transcription"])
    vocab = list(set(all_text))
    return {"vocab": [vocab], "all_text": [all_text]}

vocabs = dataset.map(
    extract_all_chars,
    batched=True,
    batch_size=-1,
    keep_in_memory=True,
    remove_columns=dataset.column_names,
)

dataset_vocab = set(vocabs["vocab"][0])
tokenizer_vocab = set(tokenizer.get_vocab().keys())
```
